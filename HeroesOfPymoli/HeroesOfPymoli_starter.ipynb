```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python

#created bins and labels
bins = [0, 10, 14, 19, 24, 29, 34, 39, 100]
binlabels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#cut df into bins
pd.cut(df["Age"], bins, labels=binlabels)
df["Age Group"] = pd.cut(df["Age"], bins, labels=binlabels)

#found total players in each age group
df_age_unique_players = df.groupby(["SN"])
value_count_for_age = df_age_unique_players["Age Group"].value_counts()
group_by_age = value_count_for_age.groupby("Age Group")
group_by_age_count = group_by_age.count()

#found age group percent
total_age = group_by_age_count.sum()
age_percent = (group_by_age_count / total_age)*100

#create final data frame
original_group_by_age = df.groupby("Age Group")
final_age_df = original_group_by_age.count()

#add new data to created data frame
final_age_df["Total Count"] = group_by_age_count
final_age_df["Percentage of Players"] = age_percent

#clean it up
final_age_df_clean = final_age_df[["Total Count", "Percentage of Players"]]
final_age_df_clean.loc[:,"Percentage of Players"] = final_age_df_clean["Percentage of Players"].astype(float).map("{:,.2f}%".format)
final_age_df_clean
```

## Top Spenders

* Run basic calculations to obtain the results in the table below

```python
#create and sort data frame
df_top_spenders = df.groupby("SN")
df_spenders_count = df_top_spenders.count() 
df_spenders_sorted = pd.DataFrame(df_spenders_count).sort_values("Purchase ID", ascending=False)
#find average and total spent
average_spent = df_top_spenders["Price"].mean()
total_spent = df_top_spenders["Price"].sum()
#add to data frame
df_spenders_sorted["Average Purchase Price"] = average_spent
df_spenders_sorted["Total Purchase Value"] = total_spent
#clean it up
df_spenders_sorted_clean = df_spenders_sorted[["Purchase ID", "Average Purchase Price", "Total Purchase Value"]]
df_spenders_sorted_clean.rename(columns={"Purchase ID": "Purchase Count"}, inplace = True)
df_spenders_sorted_clean.loc[:,"Average Purchase Price"] = df_spenders_sorted_clean["Average Purchase Price"].astype(float).map("${:,.2f}".format)
df_spenders_sorted_clean.loc[:,"Total Purchase Value"] = df_spenders_sorted_clean["Total Purchase Value"].astype(float).map("${:,.2f}".format)
#only show top 5
df_spenders_sorted_clean.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#retrieve necessary columns
df_items = df[["Item ID", "Item Name", "Price"]]
df_item_group = df_items.groupby(["Item ID", "Item Name"])
#create data frame
df_item_count = df_item_group.count()
df_pop_items = pd.DataFrame(df_item_count)
#sort Data Frame
df_pop_items_sorted = df_pop_items.sort_values("Price", ascending=False)
df_pop_items_sorted
#find price and total (using mean because it should be the same as the price)
price = df_item_group["Price"].mean()
total_value_item = df_item_group["Price"].sum()
#append data frame
df_pop_items_sorted["Item Price"] = price
df_pop_items_sorted["Total Purchase Value"] = total_value_item

#rename price
df_pop_items_sorted.rename(columns={"Price": "Purchase Count"}, inplace = True)
df_pop_items_sorted.head(5)

#dollar signs
#have to create a new df because the dollar signs mess up the sorting for the next section
df_pop_items_sorted_dollar = df_pop_items_sorted[["Purchase Count", "Item Price", "Total Purchase Value"]]
df_pop_items_sorted_dollar.loc[:,"Item Price"] = df_pop_items_sorted_dollar["Item Price"].astype(float).map("${:,.2f}".format)
df_pop_items_sorted_dollar.loc[:,"Total Purchase Value"] = df_pop_items_sorted_dollar["Total Purchase Value"].astype(float).map("${:,.2f}".format)
df_pop_items_sorted_dollar.head(5)
```

* Sort the above table by total purchase value in descending order
